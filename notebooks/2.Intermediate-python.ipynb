## Dictionaries in python
- We cannot always use list to show the related objects as it is not convenient and intuitive.

```python
# Definition of fruits and their colors
fruits = ['apple', 'banana', 'grape', 'orange']
colors = ['red', 'yellow', 'purple', 'orange']

# Get index of 'grape': ind_grape
ind_grape = fruits.index('grape')

# Use ind_grape to print out color of grape
print(colors[ind_grape])
```

- So we can clearly see, it becomes very difficult and slow to always access related objects using list. However, we can use key and value pairs using dictionary in python to store the related objects which is the most efficient way.

```python
# Definition of animals and their sounds
animals = ['dog', 'cat', 'cow', 'sheep']
sounds = ['bark', 'meow', 'moo', 'bleat']

# From string in animals and sounds, create dictionary animal_sounds
animal_sounds = {'dog': 'bark', 'cat': 'meow', 'cow': 'moo', 'sheep': 'bleat'}

# Print animal_sounds
print(animal_sounds)
```

- Using keys we can get values easily. Example is here dog is a key and bark is a value.

```python
# Definition of dictionary
programming_languages = {'python': 'easy', 'java': 'versatile', 'c++': 'powerful', 'javascript': 'interactive'}

# Print out the keys in programming_languages
print(programming_languages.keys())

# Print out value that belongs to key 'javascript'
print(programming_languages['javascript'])
```

- Dictionaries are immutable objects which means that they cannot be changed after the creation. some of other immutable objects are strings, integer, boolean etc.. and mutable objects are list and so on..
- Also we can update the value using key even after creation.
- We can also delete a key using del(dictionary[key]).
- Dictionary should be used when we are dealing with lookup values such in table so that each key is unique option.

```python
# Definition of dictionary
fruit_colors = {'apple': 'red', 'banana': 'yellow', 'grape': 'purple', 'orange': 'orange'}

# Add watermelon to fruit_colors
fruit_colors['watermelon'] = 'green'

# Print out if 'watermelon' is in fruit_colors
print('watermelon' in fruit_colors)   # this clarifies if watermelon exists in the dictionary or not

# Add blueberry to fruit_colors
fruit_colors['blueberry'] = 'blue'

# Print fruit_colors
print(fruit_colors)
```

```python
# Definition of dictionary
planet_moons = {'earth': 1, 'mars': 2, 'jupiter': 79,
                'saturn': 82, 'uranus': 27, 'neptune': 14,
                'pluto': 5}

# Update moon count of jupiter
planet_moons['jupiter'] = 80

# Remove pluto
del(planet_moons['pluto'])

# Print planet_moons
print(planet_moons)
```

### Dictionaries within Dictionary
- As list may contains lists and similar goes with dictionary where a dictionary key may contain values that is again dictionaries which is basically called nested dictionaries.

```python
# Dictionary of dictionaries
cars = {
    'tesla': { 'model': 'Model S', 'range': 396 },
    'ford': { 'model': 'Mustang', 'range': 300 },
    'toyota': { 'model': 'Prius', 'range': 600 },
    'bmw': { 'model': 'i8', 'range': 330 }
}

# Print out the range of Toyota
print(cars['toyota']['range'])

# Create sub-dictionary new_car
new_car = {'model': 'Civic', 'range': 450}

# Add new_car to cars under key 'honda'
cars['honda'] = new_car

# Print cars
print(cars)
```

```python
# Create arrays
import numpy as np
scores_a = np.array([75, 82, 91, 68])
scores_b = np.array([80, 79, 85, 65])

# scores_a greater than 80 or less than 70
print(np.logical_or(scores_a > 80, scores_a < 70))

# Both scores_a and scores_b less than 70
print(np.logical_and(scores_a < 70, scores_b < 70))
```

### Loop over DataFrame
- .iterrows() is used to loop through each row of a DataFrame.
- In each iteration, it returns:
  - lab: the row label (index).
  - row: the data of the row (as a Series).

```python
# Import pandas
import pandas as pd

# Create a small DataFrame about cities
cities = pd.DataFrame({
    'city': ['New York', 'Paris', 'Tokyo', 'Mumbai', 'Sydney'],
    'population_million': [8.4, 2.1, 9.3, 12.5, 5.3],
    'has_subway': [True, True, True, False, True]
})

# Set custom row labels
cities.index = ['NY', 'PA', 'TK', 'MU', 'SY']

# Loop over the rows
for label, row in cities.iterrows():
    print(label)
    print(row)
```

- In each run of .iterrows(), the row data is a Pandas Series.
- A Series is not very convenient to print directly.
- To access specific values, use square brackets [].

```python
# Create DataFrame about countries
countries = pd.DataFrame({
    'country': ['Nepal', 'Germany', 'Brazil', 'South Africa', 'Canada'],
    'internet_users_percent': [65.9, 89.7, 81.0, 70.0, 94.0]
})

# Set custom index labels
countries.index = ['NP', 'DE', 'BR', 'ZA', 'CA']

# Loop and print custom output
for code, row in countries.iterrows():
    print(code + ": " + str(row['internet_users_percent']) + "% internet users")
```

```python
# Create a small DataFrame about countries
countries = pd.DataFrame({
    'country': ['Nepal', 'Germany', 'Brazil', 'South Africa', 'Canada'],
    'capital': ['Kathmandu', 'Berlin', 'Braslia', 'Pretoria', 'Ottawa']
})

# Set custom index labels
countries.index = ['NP', 'DE', 'BR', 'ZA', 'CA']

# Code for loop that adds CAPITAL LENGTH column
for lab, row in countries.iterrows():
    countries.loc[lab, 'CAPITAL_LENGTH'] = len(row['capital'])

# Print the updated DataFrame
print(countries)
```

- iterrows() is easy to use but inefficient because it creates a new Pandas Series on every iteration.
- Better approach: Use apply() for column-wise operations like adding a new column or applying a function.

```python
# Create a DataFrame about countries
countries = pd.DataFrame({
    'country': ['Nepal', 'Germany', 'Brazil', 'South Africa', 'Canada'],
    'capital': ['Kathmandu', 'Berlin', 'Braslia', 'Pretoria', 'Ottawa']
})

# Set custom index labels
countries.index = ['NP', 'DE', 'BR', 'ZA', 'CA']

# Convert the capital names to uppercase
countries['CAPITAL'] = countries['capital'].apply(str.upper)

# Print the updated DataFrame
print(countries)
```

### Random Walk
- A random walk is a path where each step is determined randomly.
- At each step, you move up or down (or in different directions) based on chance.
- It is used to model unpredictable processes like stock prices, animal movements, or games.
- Example in one dimension:
- Start at position 0.
- Toss a coin:
  - Heads  move +1 step forward.
  - Tails  move -1 step backward.
- Over time, the path looks random and unpredictable.

```python
# NumPy is imported, seed is set
import numpy as np
np.random.seed(456)

# Initialize random_walk
random_walk = [5]

# Simulate the random walk
for i in range(100):
    # Current position
    position = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1, 7)

    # Define movement
    if dice == 1:
        position = max(0,position-2)
    elif dice <= 4:
        position += 1
    else:
        position += np.random.randint(1, 5)

    # Append new position
    random_walk.append(position)

# Print the result
print(random_walk)
```

- Visualize the walk

```python
import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.show()
```

- Simulate multiple walks

```python
# numpy and matplotlib imported, seed set.
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)

# Initialize and populate all_walks
all_walks = []
for i in range(5):
    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1, 7)

        # Movement logic
        if dice <= 2:
            step = max(0, step - 1)  # Prevent going below 0
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)

        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array
np_aw = np.array(all_walks)

# Plot np_aw (each walk separately)
plt.plot(np_aw)
plt.title('Random Walks - Non-Transposed')
plt.xlabel('Step Number')
plt.ylabel('Position')
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t (step-by-step evolution)
plt.plot(np_aw_t)
plt.title('Random Walks - Transposed (All Walks Together)')
plt.xlabel('Step Number')
plt.ylabel('Position')
plt.show()
```

- Plot distribution

```python
np.random.seed(123)

# Simulate random walk 500 times
all_walks = []
for i in range(500):
    random_walk = [0]
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1, 7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step += 1
        else:
            step += np.random.randint(1, 7)

        # 0.1% chance to fall back to start
        if np.random.rand() <= 0.001:
            step = 0

        random_walk.append(step)

    all_walks.append(random_walk)

# Create and transpose array
np_aw_t = np.transpose(np.array(all_walks))

# Select last row: ends
ends = np_aw_t[-1, :]

# Plot histogram
plt.hist(ends, bins=20, edgecolor='black')
plt.title('Distribution of Final Steps After 100 Moves')
plt.xlabel('Final Position')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

- From this graph we can clearly see it is like bell shaped like in normal graphical distribution.
