## Dictionaries in python
- We cannot always use list to show the related objects as it is not convenient and intuitive.

```python
# Definition of fruits and their colors
fruits = ['apple', 'banana', 'grape', 'orange']
colors = ['red', 'yellow', 'purple', 'orange']

# Get index of 'grape': ind_grape
ind_grape = fruits.index('grape')

# Use ind_grape to print out color of grape
print(colors[ind_grape])
```

- So we can clearly see, it becomes very difficult and slow to always access related objects using list. However, we can use key and value pairs using dictionary in python to store the related objects which is the most efficient way.

```python
# Definition of animals and their sounds
animals = ['dog', 'cat', 'cow', 'sheep']
sounds = ['bark', 'meow', 'moo', 'bleat']

# From string in animals and sounds, create dictionary animal_sounds
animal_sounds = {'dog': 'bark', 'cat': 'meow', 'cow': 'moo', 'sheep': 'bleat'}

# Print animal_sounds
print(animal_sounds)
```

- Using keys we can get values easily. Example is here dog is a key and bark is a value.

```python
# Definition of dictionary
programming_languages = {'python': 'easy', 'java': 'versatile', 'c++': 'powerful', 'javascript': 'interactive'}

# Print out the keys in programming_languages
print(programming_languages.keys())

# Print out value that belongs to key 'javascript'
print(programming_languages['javascript'])
```

- Dictionaries are immutable objects which means that they cannot be changed after the creation. some of other immutable objects are strings, integer, boolean etc.. and mutable objects are list and so on..
- Also we can update the value using key even after creation.
- We can also delete a key using del(dictionary[key]).
- Dictionary should be used when we are dealing with lookup values such in table so that each key is unique option.

```python
# Definition of dictionary
fruit_colors = {'apple': 'red', 'banana': 'yellow', 'grape': 'purple', 'orange': 'orange'}

# Add watermelon to fruit_colors
fruit_colors['watermelon'] = 'green'

# Print out if 'watermelon' is in fruit_colors
print('watermelon' in fruit_colors)   # this clarifies if watermelon exists in the dictionary or not

# Add blueberry to fruit_colors
fruit_colors['blueberry'] = 'blue'

# Print fruit_colors
print(fruit_colors)
```

```python
# Definition of dictionary
planet_moons = {'earth': 1, 'mars': 2, 'jupiter': 79,
                'saturn': 82, 'uranus': 27, 'neptune': 14,
                'pluto': 5}

# Update moon count of jupiter
planet_moons['jupiter'] = 80

# Remove pluto
del(planet_moons['pluto'])

# Print planet_moons
print(planet_moons)
```

### Dictionaries within Dictionary
- As list may contains lists and similar goes with dictionary where a dictionary key may contain values that is again dictionaries which is basically called nested dictionaries.

```python
# Dictionary of dictionaries
cars = {
    'tesla': { 'model': 'Model S', 'range': 396 },
    'ford': { 'model': 'Mustang', 'range': 300 },
    'toyota': { 'model': 'Prius', 'range': 600 },
    'bmw': { 'model': 'i8', 'range': 330 }
}

# Print out the range of Toyota
print(cars['toyota']['range'])

# Create sub-dictionary new_car
new_car = {'model': 'Civic', 'range': 450}

# Add new_car to cars under key 'honda'
cars['honda'] = new_car

# Print cars
print(cars)
```

## Pandas for Data Science
- Different ways of handling datasets in python. You may think of using 2d numpy array but it is used for only single datatype and we need to handle multiple datatypes in table when working with datasets.
- So we basically use pandas for datasets. Pandas is a high level data manipulation tool that is built on top of numpy.
- In pandas we store the tabular data in a object called the dataframe.
### DataFrame from Dictionary

```python
# Pre-defined lists
languages = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby', 'Go', 'Swift']
is_interpreted = [True, False, False, True, True, False, False]
popularity_index = [1, 2, 4, 3, 10, 7, 6]

# Import pandas as pd
import pandas as pd

# Create dictionary lang_dict with three key:value pairs
lang_dict = {'language': languages, 'interpreted': is_interpreted, 'popularity_rank': popularity_index}

# Build a DataFrame langs from lang_dict
langs = pd.DataFrame(lang_dict)

# Print langs
print(langs)
```

```python
# Definition of row_labels
row_labels = ['PY', 'JA', 'C++', 'JS', 'RB', 'GO', 'SW']

# Specify row labels of langs
langs.index = row_labels

# Print langs again
print(langs)
```

### Index and Select Data
- Numerous ways to index and select data from dataframe:
  - square brackets.
  - advanced methods such as loc and iloc.
- Column Access:
  - langs[["is_interpreted"]]
  - langs[["is_interpreted" , "popularity_rank"]]
- Row Access:
  - langs[1:4]

- These square brackets has limited functionality.
- Ideally what we need is like 2D numpy arrays which is in the form of rows and columns.
- So for pandas we use loc(label based) and iloc(integer-position based).

- Row Access loc:
  - langs.loc[["Python" , "Java"]]

- Row and Column loc:
  - langs.loc[["Python" , "Java"], ["is_interpreted"]]

- Row Access iloc:
  - langs.iloc[1]

- Row and Column iloc:
  - langs.iloc[[1,2], [0,1]]

```python
# Import languages data
import pandas as pd
langs = pd.read_csv('../resources/files/languages.csv', index_col=0)

# Print out language column as Pandas Series
print(langs["language"])

# Print out language column as Pandas DataFrame
print(langs[["language"]])

# Print out DataFrame with language and is_interpreted columns
print(langs[["language", "is_interpreted"]])
```

```python
# Print out first 3 observations
print(langs[0:3])

print("\n")

# Print out fourth, fifth and sixth observation
print(langs[3:6])
```

```python
# Print out observation for JavaScript
print(langs.loc["JS"])
print("\n")
print(langs)
print("\n")

# Print out observations for Java and Swift
print(langs.iloc[[1, 6]])
```

```python
# Print out is_interpreted value of Ruby
print(langs.loc[["RB"], ["is_interpreted"]])

print("\n")

# Print sub-DataFrame for Go and Swift with columns language and popularity_rank
print(langs.loc[["GO", "SW"], ["language", "popularity_rank"]])
```

```python
# Print out is_interpreted column as Series
print(langs["is_interpreted"])
print("\n")

# Print out is_interpreted column as DataFrame
print(langs.loc[:, ["is_interpreted"]])

print("\n")
# Print out language and is_interpreted columns as DataFrame
print(langs.loc[:, ["language", "is_interpreted"]])
```

## Boolean Operators
- For 1D values we use and, or and not operatiors but for numpy arrays we have to use logical_and, logical_or and logical_not.

```python
# Define variables
room1 = 22.5
room2 = 15.0

# room1 bigger than 20 and smaller than 25?
print(room1 > 20 and room1 < 25)

# room1 smaller than 10 or bigger than 18?
print(room1 < 10 or room1 > 18)

# Double room2 smaller than triple room1?
print(room2 * 2 < room1 * 3)
```

```python
# Create arrays
import numpy as np
scores_a = np.array([75, 82, 91, 68])
scores_b = np.array([80, 79, 85, 65])

# scores_a greater than 80 or less than 70
print(np.logical_or(scores_a > 80, scores_a < 70))

# Both scores_a and scores_b less than 70
print(np.logical_and(scores_a < 70, scores_b < 70))
```

## Conditional statements

```python
# Define variables
space = "bed"
size = 18.5

# if statement for space
if space == "bed":
    print("This is the bedroom.")

# if statement for size
if size > 20:
    print("Spacious room!")
```

```python
# Define variables
space = "bath"
size = 12.0

# if-else construct for space
if space == "bath":
    print("checking out the bathroom.")
else:
    print("checking out a different room.")

# if-else construct for size
if size >= 15:
    print("quite spacious!")
else:
    print("a bit cramped.")
```

```python
# Define variables
room = "bath"
area = 19.5

# if-elif-else construct for room
if room == "kit":
    print("This is the kitchen.")
elif room == "bath":
    print("This is the bathroom.")
else:
    print("Some other room.")

# if-elif-else construct for area
if area > 20:
    print("Very spacious!")
elif area > 15:
    print("Comfortable size.")
else:
    print("Tiny spot!")
```
