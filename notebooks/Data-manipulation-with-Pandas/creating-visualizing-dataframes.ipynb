# Creating and Visualizing DataFrames
## Visualizing data

```python
# üì¶ Step 1: Import matplotlib.pyplot with alias plt
import matplotlib.pyplot as plt
import pandas as pd

coffee_sales = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\KaranCodes\Datacampcourses\Associate-Data-Scientist-Python-Track\resources\DatamanipulationwithPandas-datasets\coffee_sales.csv")

# üëÄ Step 2: Print the first few rows of the dataset
print(coffee_sales.head())

# üìä Step 3: Calculate total coffee units sold for each size
nb_sold_by_size = coffee_sales.groupby("size")["nb_sold"].sum()

# üñºÔ∏è Step 4: Create a bar plot of the number of units sold by size
nb_sold_by_size.plot(kind="bar", color="peru", title="Total Coffee Units Sold by Size")

# ‚ú® Step 5: Label the axes
plt.xlabel("Coffee Size")
plt.ylabel("Number of Units Sold")

# üìà Step 6: Show the plot
plt.tight_layout()
plt.show()
```

### Changes in sales over time
- Line plots are ideal for showing the relationship between two numeric variables.
- Each data point is connected to the next, forming a continuous line.
- Particularly useful for:
  - Visualizing trends over time.
  - Tracking changes in a value across time points.
- Common use case: tracking sales, stock prices, temperatures, etc.

```python
# Calculate total coffee units sold on each date
nb_sold_by_date = coffee_sales.groupby("date")["nb_sold"].sum()

# Create a line plot of the number of units sold by date
nb_sold_by_date.plot(kind="line", color="saddlebrown", title="Coffee Units Sold Over Time",rot=45)

# Label the axes
plt.xlabel("Date")
plt.ylabel("Units Sold")

# Display the plot
plt.tight_layout()
plt.show()
```

### Coffee supply and demand
- Scatter plots are great for visualizing relationships between two numerical variables.
- Each point represents a pair of values (e.g., one x-value and one y-value).
- Useful for:
  - Spotting correlations or patterns.
  - Determining if one variable can help predict another.
- Example: comparing avocados sold vs. average price to explore supply and demand dynamics.

```python
# Scatter plot of avg_price vs. nb_sold with title
coffee_sales.plot(x="nb_sold", y="avg_price", title="Number of coffees sold vs. average price", kind="scatter")

# Show the plot
plt.show()
```

### Layering plots
- Plots can also be layered on top of one another.

```python
# Filter for Arabica and Robusta types and plot histograms of avg_price with bins = 20
coffee_sales[coffee_sales["type"] == "arabica"]["avg_price"].hist(alpha=0.5, bins=20, label="Arabica")
coffee_sales[coffee_sales["type"] == "robusta"]["avg_price"].hist(alpha=0.5, bins=20, label="Robusta")

# Add a legend
plt.legend()    #plt.legend(["Arabica", "Robusta"])

# Add titles and labels
plt.title('Average Price Distribution of Arabica vs Robusta Coffees')
plt.xlabel('Average Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
