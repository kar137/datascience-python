## Matplotlib
- In python we have many visualization packages but the mother of all we can say is matplotlib.
- We need it's subpackage pyplot

```python
#visualizing evolution of world's population
import matplotlib.pyplot as plt 
year = [1950, 1970 , 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, population)  # plt.plot(horizontal_axis, vertical_axis)  
plt.show()
```

- plot function tells python what to plot and how to plot whereas show displays the plot.
### Scatter Plot
- It helps to simply plot all invidual data points and it helps to clearly show how many points the plot are based on.

```python
plt.scatter(year, population)  # plt.scatter(horizontal_axis, vertical_axis)  
plt.show()
```

```python
# Sample data for months and sales
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [1200, 1500, 1800, 2100, 2500, 2300, 2700, 3000, 2800, 3200, 3500, 4000]

# Print the last item from months and sales
print(months[-1])  # Last month
print(sales[-1])   # Last month's sales

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Make a line plot: months on the x-axis, sales on the y-axis
plt.plot(months, sales)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data: advertising budget (in $1000) vs sales revenue (in $1000)
ad_budget = [1, 5, 10, 50, 100, 500, 1000, 5000, 10000]  # Advertising budget
sales_revenue = [5, 20, 35, 80, 150, 400, 700, 3000, 6000]  # Corresponding sales revenue

# Create a scatter plot
plt.scatter(ad_budget, sales_revenue)

# Put the x-axis on a logarithmic scale
plt.xscale('log')

# Show plot
plt.show()
```

## Histogram
- A type of visualization useful for exploring dataset.
- It helps to get idea about distribution of variables.
### Using matplotlib for histogram

```python
import matplotlib.pyplot as plt
help(plt.hist)
```

- x argument is the list of values that we want to build the histogram for and bins argument is used to tell for how many bins the data should be divided and by default it is 10.

```python
values = [1, 1.1, 2.6, 3.5, 1.3, 0.5, 0.9, 1.9, 2.3, 3.2, 3.4, 2.1]
plt.hist(values, bins=3)
plt.show()
```
